```python
size = 128
bi_matri = []
for linha in img:
    bi_matri.append([celula[0] for celula in linha])
```

```python
np.array(bi_matri)
```

```python
def pretty_print_bi_matri(img: np.array):
    res = ''
    for linha in img:
        for p in linha:
            res += f'{int(p):>1}'
        res += '\n'
    print(res)

```

```python
pretty_print_bi_matri(bi_matri)
```
